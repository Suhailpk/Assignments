```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Q9_a.csv')
df
```

```python
df.skew()
```

```python
df.kurtosis()
```

```python
sns.distplot(df.speed, hist=False)
plt.title('SPEED')
plt.show()
```

```python
sns.distplot(df.dist, hist=False)
plt.title('DISTANCE')
plt.show()
```
